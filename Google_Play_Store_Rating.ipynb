# Key Features of the Dataset:

App Information: The dataset includes information about different apps, such as their names, categories, and genres, providing a comprehensive overview of the app landscape.

Rating: App ratings are provided, which reflect user feedback and satisfaction levels. Ratings are essential indicators of app quality and user experiences.

Reviews: The number of reviews for each app is included, allowing for an assessment of user engagement and popularity.

Size: App sizes in terms of storage are provided, offering insights into the space requirements of different apps.

Installs: Information about the number of app installations provides an indication of app popularity and user reach.

Type: The type of app (free or paid) is specified, aiding in the analysis of revenue models and user preferences.

Price: For paid apps, the pricing information is included, facilitating analyses related to pricing strategies and user willingness to pay.

Content Rating: The content rating of the app is provided, helping to understand the target audience and app suitability for different age groups.

Genres: The dataset includes information about the genres to which each app belongs, allowing for genre-specific analyses.

Last Updated: The date of the app's last update is included, which can be useful for tracking app development and maintenance.

Current Version: The current version of each app is specified, aiding in version-specific analyses.

Android Version: The minimum required Android version for each app is provided, assisting in compatibility assessments.

## **Remove Columns that are 90% Empty**

```python
threshold = len(df) * 0.1   # 0.1 stands for 10 % means if columns having only 10% value filled and 90 % Empty just drop that column
threshold
```

```python
df.dropna(thresh= threshold, axis = 1, inplace= True)
```

```python
print(df.isnull().sum())  # So there I got no any changes in the NULL values so I have to go for another approach to handle the missing values
```

# **Data Manipulation**

```python
# for Numerical Values/column = Fill with Median of that column
# for Categorical Values/column = Fill with Mode of that column


def impute_median(series):
  return series.fillna(series.median)     # 'fillna' is a function which is used to fill the particular column
```

```python
df.Rating = df['Rating'].transform(impute_median)
```

```python
df.isnull().sum()
```

So Yes I have removed the NUll values from Rating Columns by Imputing with its Median Value Using function I made 'impute_median' and it Removed the NULL Values. This process is called DATA MANIPULATION

```python
print(df['Type'].mode())
print(df['Current Ver'].mode())
print(df['Android Ver'].mode())
```

```python
df['Type'].fillna(str(df['Type'].mode().values[0]),inplace=True)
df['Current Ver'].fillna(str(df['Current Ver'].mode().values[0]),inplace=True)
df['Android Ver'].fillna(str(df['Android Ver'].mode().values[0]),inplace=True)
```

```python
df.isnull().sum()
```

```python
df['Price'] = df['Price'].apply((lambda x: str(x).replace('$','') if '$' in str (x) else str(x)))
df['Price'] = df['Price'].apply(lambda x: float(x))

df['Reviews'] = pd.to_numeric(df['Reviews'],errors = 'coerce')
```

```python
df['Installs'] = df['Installs'].apply(lambda x: str(x).replace('+','') if '+' in str(x) else str(x))
df['Installs'] = df['Installs'].apply(lambda x: str(x).replace(',','') if ',' in str(x) else str(x))
df['Installs'] = df['Installs'].apply(lambda x: float(x))
```

## **DATA VISUALIZATION**

```python
# Here I am grouping the Installs,Price,Reviews Columns according to the Category means
# for each category calculating (Mean of Installs, Sum of Price, Mean of Reviews)

grp = df.groupby('Category')
x = grp['Installs'].agg(np.mean)
y = grp['Price'].agg(np.sum)
z= grp['Reviews'].agg(np.mean)
print(x)
print(y)
print(z)
```

```python
plt.figure(figsize = (16,5))
plt.plot(x, 'ro' ,  color = 'blue')    # 'ro' given to show the ROUND circles
plt.xticks(rotation = 90)              # rotation is given to not messed up the x-axis column name so i rotated the text by 90 degree for clear visuals
plt.title('Category vs Installs')      # as above i stored x-axis as 'Installs'
plt.xlabel('Categories -------->')
plt.ylabel('Installs---------->')
plt.show()
```

```python
plt.figure(figsize = (16,5))
plt.plot(y, 'r--', color = 'red')    # 'r--' given to show the  DASHED LINES
plt.xticks(rotation=90)              # roatated the text on X axis by 90 degree for clear Visuals
plt.title('Category vs Pricing')      # as above i have already stored the y axis as 'Pricing'
plt.xlabel('Categories -------->')
plt.ylabel('Prices ---------->')
plt.show()

```

```python
plt.figure(figsize=(16,5))
plt.plot(z, 'g^' , color='green')    # g^ uses for 'triangle' shape plot
plt.xticks(rotation=90)
plt.title('Category vs Reviews')
plt.xlabel('Category---------->')
plt.ylabel('Reviews------------>')
plt.show()

```
